```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/Rush/.spyder-py3/flairminds/dataset/diamonds.csv")
df.head()
```

```python
df['cut'].unique()
```

```python
cut_class_dict = {"Fair":1, "Good":2, "Very Good":3, "Premium":4, "Ideal":5}
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J":1, "I":2, "H":3, "G":4, "F":5, "E":6, "D":7}

df['cut'] = df['cut'].map(cut_class_dict)
df['clarity'] = df['clarity'].map(clarity_dict)
df['color'] = df['color'].map(color_dict)
df.head()
```

```python
# # For converting NUM values back into imeaningful info, swap key,values:
# new_dict = {v: k for k, v in d.items()}
```

```python
import sklearn
from sklearn import svm, preprocessing

df = sklearn.utils.shuffle(df)

x = df.drop("price", axis=1).values
x = preprocessing.scale(x)
y = df['price'].values

test_size =200

x_train = x[:-test_size]
y_train = y[:-test_size]

x_test = x[-test_size:]
y_test = y[-test_size:]

clf = svm.SVR(kernel="linear")
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
for x,y in zip(x_test, y_test):
    print(f"Model: {clf.predict([x])[0]}, Actual: {y}")
# six negative predictions LOL
```

```python
#new model
clf = svm.SVR(kernel="rbf")
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
for x,y in zip(x_test, y_test):
    print(f"Model: {clf.predict([x])[0]}, Actual: {y}")
# poor accuracy ..at least no negative price
```
