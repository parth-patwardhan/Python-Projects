```python
df_graph = pd.DataFrame()

for region in df.region.unique()[:10]:
    df_region = df.copy()[df.region == region]
    df_region.set_index("Date", inplace=True)
    df_region.sort_index(inplace=True)
    df_region[f"{region}_RM"] = round(df_region["AveragePrice"].rolling(25).mean(),2)
    
    if df_graph.empty:
        df_graph = df_region[[f"{region}_RM"]]
    else:
        df_graph = df_graph.join(df_region[f"{region}_RM"])
```
